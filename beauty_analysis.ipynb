## The Top 10 **Beauty Brands** with most products on the 2024 Top

```python
brand = pd.DataFrame(beauty_df['Brand'].value_counts()).reset_index()
brand = brand[:10]
```

```python
# General view
plt.figure(figsize = (7,6))
sns.barplot(brand, x = 'Brand', y = 'count', palette = 'Paired')
plt.title('Top 10 Beauty Brands with most products on the 2024 Top')
plt.xlabel('Brands')
plt.xticks(fontsize = 8, rotation = 45)
plt.ylabel('Count')
plt.show();
```

```python
# Zoom in
plt.figure(figsize = (7,6))
sns.barplot(brand, x = 'Brand', y = 'count', palette = 'Paired')
plt.title('Top 10 Beauty Brands with most products on the 2024 Top')
plt.xlabel('Brands')
plt.xticks(fontsize = 8, rotation = 45)
plt.ylabel('Count')
plt.ylim(bottom=350, top = 430)
plt.show();
```

## The Most Popular **Beauty Product Category** in 2024
